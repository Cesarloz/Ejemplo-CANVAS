```
#matriz1=input("matriz")
from sympy import *
import numpy as np

import scipy.linalg as la

var0=[[1,4,5],[4,6,1],[9,3,6]]

var = np.array(var0,float)



def rrenglon(f,k,v):
  var[k+1:k+2]=(v*var[k+1,f]-var[k+1:k+2])


def mrenglon(f,k):
  if(var[f,f]!=0):
    var[f:k]=var[f:k]*(1/var[f,f])
  else:
    var[f:k]=var[f:k]

def mgrenglon(f,k):
  var[i:i+1]=(-var[j:j+1]*var[i,j]+var[i:i+1])


c,f=var.shape
print(c)
print(f)
I=np.eye(c)
if(c==f):
  print("la matriz es cuadrada")
else:
  print("la matriz no es cuadrada")


var= np.hstack([var,I])
print("\n",var)



#gauss 
l=0
for i in range(0,c):
  l+=1
  mrenglon(i,l)
  print(var)
  for j in range(i,(c-1)):
     rrenglon(i,j,var[i:i+1])
     print(var)

#Jordan
for j in range((c-1),0,-1):
  print("este",j)
  for i in range((j-1),-1,-1):
    print("otro",i)
    mgrenglon(i,j-1)
    print(var)



print("\n",var)

print("l \n",var)


print("La funcion inversa con linalg \n {}".format(la.inv(var0)))
```
